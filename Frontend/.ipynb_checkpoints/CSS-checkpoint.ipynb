#                                            combining CSS selector

1.group selector ----->>> used to apply same styles to multiple elements see how

    color: blue;
}
mtlb h1,h2 and p element ka font color red hona chahiye simple



2.child selector ----->>> used to target all the direct childs(only one level nested elements) of a parent but not to parent see

ul > li > a{
    color:red
}

mtlb ul element ke ander jaao aur phir ul ke direct li element ke ander jao aur phir li ke direct a element ko target kro simple ya essi ko yu kahu ki ul ke direct li child ke direct a child ko style krna hai simple 


3.descendent selector ----->>>  is a selector that matches all elements that are descendants of a specified element

div p {
    color: blue;
}


4.chaining selector ----->>> selectors without space see

div.hlo#hii{
    color : red
}
mtlb wah element jo ki div hona chahiye aur uske pass hlo and hii name ki class aur id honi chhaiye to usko ye style apply ho rhi hogi simple



5.Adjacent sibling selector (+): kisi ek element  ke just baad waala eleement select krna ho then use this but both should be                                    the childs of same parent

6.Adjacent general selector (~): kisi ek element ke baad waale saare eleements ko target krna ho then use this

justify-conetent:

used to align/arrange flex-items together along main axis
possible values for justify-content
a.flex-start
b.flex-end
c.space-between
d.space-around
e.space-evenly
f.center

align-items:

used to align/arrange the flex-items together along the cross axis if enough space is available aling cross axis
possible values for align-itmes
a.flex-start
b.flex-end
c.center
d.baseline
e.stretch

align-self:

used to align any individual flex item along the cross axis aur yah align-item ya align-content ya justify-content ko overeride krdeta hai
possible values for it

a.flex-start
b.flex-end
c.center
d.baseline
e.stretch


                                                      FLEX SIZING
                                                      
                                              (mtlb how to size flex-itmes)
                                        
ab chunki kisi bhi elements ka size/dimension (height and width) set/control krne ke liye hm height and width property ka istemaal krte hai but in case of flex-items (jo ki elements hi hote hai) height and width property ke saath saath kuchh aur properties bhi use kr skte hai to control the size/dimension of flex-items jaise flex-basis and min/max-width/height see the priority order

conten-width < width < flex-basis < min-width/max-width
content-height < height < flex-basis < min-height/max-height

                                              WIDTH/HEIGHT VS FLEX-BASIS
                                      
             koi fark nhi padta ki kiski value chhoti ya badi hai hmesha flex-basis ki value hi implement hogi
           
           
                                      WIDTH/HEIGHT/FLEX-BASIS VS MIN-WIDTH/HEIGHT
                                 
{
    width/height/flex-basis:100px
    min-width/height:200px
    
    ess case me initial size of flex-items will be 100px and flex-items can shrink to 200px only
}

{
    width/height/flex-basis:200px
    min-width/height:100px
    
    ess case me initial size of flex-items will be 100px and flex-items can shrink to 100px hi only
}
                               
 
 
NOTE:

1.content-width koi property nhi hoti hai yah bs itna bta rhi hai ki agr initial size set nhi kiya gya hai by using other four     properties then by default flex-items ka initial size unke content-width/height ke brabar hoga 


2.ab jaisa ki hm dekh paa rhe hai ki flex-basis ek aisi property hai jisse height aur width dono control kr skte hai but at a     time for any flex container ya to wo height ko control kregi ya phir width ko 

  flex-basis hmesha main-axis ke along hi size ko control krti hai so
  a.if X-axis is the main axis then flex-basis width dimension ko set krne ke liye hogi
  b.if Y-axis is the main axis then flex-basis height dimension ko set krne ke liye hogi/kr skte hai 
  

3.each flex-items have ability to shrink and grow to min-width/heigth and max-width/height along main axis aur agar min-           width/height specify nhi hai tab to content-width/height aur agar max-width/height specify nhi hai tab flex-items grow till     jabtak container ka space equally distribute na  ho jaaye to all flex-items. we can on/off the shrik and grow ability of         any flex-item

#                                                         GRID


A grid is a network of intersecting parallel and  perpendicular lines that form a set of squares or rectangles where each square or rectangle known  as cell. aur phir ess pure structure ko grid bolte hai see the grid structure below

+-------+-------+-------+-------+
|       |       |       |       |
|       |       |       |       |
+-------+-------+-------+-------+
|       |       |       |       |
|       |       |       |       |
+-------+-------+-------+-------+
|       |       |       |       |
|       |       |       |       |
+-------+-------+-------+-------+
|       |       |       |       |
|       |       |       |       |
+-------+-------+-------+-------+

1.the complete horizontal path/space between two parallel lines ------>> row/ horizontal track
2.the complete horizontal path/space between two parallel lines ------>> column/ vertical track 

aaao ab samajhte hai ki actaully CSS grid ke case me kya hota hai 

1. display : grid ----->>     jaise hi hm yh property kisi container/div ko set krte hai ek non visible grid structure taiyar ho                               jaata hai ya yu kahu ki wah div/container ek grid structure me convert ho jaata hai simple

2. grid-template-row ---->    enn do property ke help se hm ye bta rhe hote hai ki hme kis size ke kitne rows and columns          grid-template-column       chahiye ya use krna chahte hai 

3. gap              ---->     ess property ki help se hm grid line ko wider/thicker bna skte hai aur jab hm aisa krte hai tab                                 grid items ke bich me ek visible gap dikhne lgta hai ya yu kahu ki if we want gap between rows and                               column then use this gap property ya yu kahu ki rows and columns jo ki ek dusre se attached/chipke                               hue the jinki wah se har ek cell ek dusre se chipke/attached the jinki wah se grid items ek dusre                               se chipke hue nazar aa rhe honge jaise hi hm ess gap property ko set krte hai waise hi ab wo rows                               and columns ya yu kahu wo cells ya wo grid items ab alag alag ho chuke honge/alag alag nazar aa                                 rhe honge mtlb finbally ab unme ek visible gap aagya hai simple

3. cell             ---->     each square or rectangle of that grid structure known as cell 

4. grid-item        ---->     our actual content(jo hm user ko display krana chahte hai).har ek grid-item cell me hote hai ek                                 grid item ek ya ek se adhik cell me ho skta hai ya yu kahu ki ek grid-item ek ya ek se adhik cell-                               space jitna space me ho skta hai ya space le skta hai 

5. grid structure   ---->     non-visible hote hai but by using chrome developer tool we can see them agr kuchh visible hote hai    and cell                   to wo hai grid-items and gap

6. rows values      ---->    rows values hme ya batati hai ki row/horizontal track ki vertical thickness(height) kya hongi
7. column values    ---->    rows values hme ya batati hai ki column/vertical track ki horizontal thickness(width) kya hongi

NOTE:

1.grid-items left to right arrange ho rhe hote hai into our grid structure 

2.ab chunki grid system ka kewl grid items and gap hi visible hote hai so isliye jaruri nmhi ki jitne hme cell ki need hai ussi   ke according hm grid structrure taiyar krenge balki hm jitne chahe utne rows and columns create kr skte hai aur phir apne       hisaab se kisi bhi grid item kahi bhi bole to kisi bhi cell me place kr skte hai aur phir hm cell ka size define kr skte hai

4.jabhi hmaare pass ek complex layout ho tab hme drne ki jaraurat nhi hai blki grid items se jyada cell/rows/columns waala grid   defined kr lena hai aisa krne se hmare pass bahut jyaada cell honge aur phir hm ek grid item ko kahi bhi place kr skte hai aur   kitne bhi cell de skte hai aur esse bhi agar baat nhi bani to cell/grid item ko resize bhi kr skte hai aur grid item ka order   bi define kr skte hai. chizo ko aur bhi acche se samajhne ke liye use chrome dev tool and see the grid for better               visualization go  and click on layout opt
